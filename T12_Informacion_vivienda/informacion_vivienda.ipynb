#### 5. Investigación terminológica

Es importante investigar y comprender los siguientes términos, que le ayudarán a comprender la naturaleza y las características de sus datos:

- Curtosis: La curtosis es una característica de forma de una distribución estadística o de probabilidad que indica el grado de concentración de los valores alrededor de la media y la forma de las colas de la distribución. 
Source: https://www.questionpro.com/blog/es/curtosis/

- Asimetría: La asimetría en estadística es una medida que indica el grado de simetría o falta de simetría de una distribución de datos respecto a su media aritmética.
Source: https://www-scribbr-co-uk.translate.goog/stats/skewness-meaning/?_x_tr_sl=en&_x_tr_tl=es&_x_tr_hl=es&_x_tr_pto=tc

#### 6. Comprobación de la distribución

```python
# importamos searbon
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Graficamos la distribución de la variable 'SalePrice'
sns.histplot(data['SalePrice'], kde=True)
plt.savefig('saleprice_distribution_normal.png', dpi=300)
```

```python
# Calculamos la curtosis y asimetría de la variable 'SalePrice'
print(f"Curtosis: {data['SalePrice'].kurtosis()}, Asimetría: {data['SalePrice'].skew()}")
```

```python
# Transformamos la variable 'SalePrice' usando logaritmo natural
log_price = np.log(data['SalePrice'])
```

```python
# Graficamos la distribución de la variable transformada 'log_price'
sns.histplot(log_price, kde=True)
plt.savefig('saleprice_distribution_log.png', dpi=300)
```

```python
# Calculamos la curtosis y asimetría de la variable transformada 'log_price'
print(f"Curtosis: {log_price.kurtosis()}, Asimetría: {log_price.skew()}")
```

#### 7. Comprobación del coeficiente de correlación

Verifique la correlación entre cada característica y determine qué variables de los datos están estrechamente relacionadas con la variable objetivo

```python
# Nos aseguramos de que los datos ean numericos antes de calcular la correlación
data = data.select_dtypes(include=[np.number])

# Calculando la correlación de person
corr = data.corr()
```

```python
# Visualizamos la matriz de correlación
plt.figure(figsize=(15, 14))
sns.heatmap(corr, cmap='coolwarm', square=True, annot=True, fmt=".1f", annot_kws={"size": 7}, cbar_kws={"shrink": .8})
plt.title('Matriz de correlación de Pearson', fontsize=14, color='darkred')
plt.savefig('correlation_matrix.png', dpi=400)
plt.show()
```

```python
# Visualizamos las 10 variables más correlacionadas con 'SalePrice'
top10_corr = corr.nlargest(10, 'SalePrice')['SalePrice'].index
plt.figure(figsize=(10, 9))
sns.heatmap(data[top10_corr].corr(), annot=True, cmap='coolwarm', square=True)
plt.title('Top 10 variables más correlacionadas con SalePrice', fontsize=14, color='darkred')
plt.savefig('top10_correlation_with_saleprice.png', dpi=400)
plt.show()
```
