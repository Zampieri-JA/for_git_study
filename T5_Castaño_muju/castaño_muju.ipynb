#### El día en que los bollos de castañas cubran el sistema solar
¿cuántos minutos tardará todo el sistema solar en cubrirse de panecillos?

```python
# Importamos librerías necesarias
import math
import matplotlib.pyplot as plt
```

Para resolver este problema, lo abordo mediante una función que me permite calcular el volumen total de los bollos en cada intervalo de tiempo (5 minutos) hasta que supera el volumen del sistema solar (aunque no se tiene un dato oficial del radio del sistema solar, decidí usar la distancia aproximada hasta el planeta neptuno que es de unos 4.500 millones de km)
Para estimar el volumen, usé esta formula:



```python
# Creamos una función para calcular el crecimiento exponencial
def simular_crecimiento(volumen_inicial, volumen_objetivo, tiempo_duplicacion_minutos):
    """
    Simula el crecimiento exponencial hasta alcanzar el volumen del objetivo.
    """
    volumen_actual = volumen_inicial
    minutos_transcurridos = 0
    volumenes_en_el_tiempo = [volumen_inicial]

    while volumen_actual < volumen_objetivo:
        volumen_actual *= 2
        minutos_transcurridos += tiempo_duplicacion_minutos
        volumenes_en_el_tiempo.append(volumen_actual)

    return minutos_transcurridos, volumenes_en_el_tiempo
```

```python
# Datos iniciales
radio_bollo = 0.02  # metros
volumen_bollo = (4/3) * math.pi * (radio_bollo**3)

# Radio estimado para el sistema solar: 4.500 millones de km 
radio_sistema_solar = 4.5 * 10**12 # metros
volumen_sistema_solar = (4/3) * math.pi * (radio_sistema_solar**3)

tiempo_duplicacion = 5  # minutos
```

```python
# Ejecutamos la función
tiempo_total, vol_historial = simular_crecimiento(volumen_bollo, volumen_sistema_solar, tiempo_duplicacion)
```

```python
# Imprmimimos los resultados
print(f"Volumen de un bollo: {volumen_bollo:.2e} m^3")
print(f"Volumen del sistema solar: {volumen_sistema_solar:.2e} m^3")
print(f"El sistema solar se cubriría en aproximadamente {tiempo_total} minutos.")
print(f"Eso es aproximadamente {tiempo_total / 60:.0f} Horas")
```

```python
# grafica del crecimiento
plt.figure(figsize=(10, 6))
plt.plot(range(0, len(vol_historial) * tiempo_duplicacion, tiempo_duplicacion), vol_historial, marker='o', linestyle='-')
plt.title('Crecimiento del Volumen de Bollos de Castaño Manju', fontsize=12, fontweight='bold', color='darkgreen')
plt.xlabel('Tiempo en Minutos', fontsize=10, fontweight='bold', color='darkgreen')
plt.ylabel('Volumen Total (m³)', fontsize=10, fontweight='bold', color='darkgreen')
plt.yscale('log') # Escala logarítmica para el crecimiento exponencial
plt.grid(True, which="major", ls="--")
plt.axhline(y=volumen_sistema_solar, color='red', linestyle='--', label='Volumen del Sistema Solar', linewidth=2)
plt.legend()
plt.savefig('castaño_manju.png', dpi=400)
plt.show()
```
