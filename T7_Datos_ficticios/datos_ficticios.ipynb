#### 1. Creación de números aleatorios

Genere 500 números aleatorios a partir de una distribución normal bidimensional con media (-3, 0) y matriz de covarianza

```python
# Importamos las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definimos la semilla
np.random.seed(0)

# Definimos la matriz de covarianza
matriz_covarianza = np.array([[1, 0.8], [0.8, 1]])

# Definimos la media
media = np.array([-3, 0])

# Generamos los datos
datos = np.random.multivariate_normal(media, matriz_covarianza, size=500)

# verificamos las dimensiones de los datos
print("\nDimensiones de los puntos aleatorios:", datos.shape)
```

#### 2. Visualización mediante un diagrama de dispersión

```python
# configuramos el gráfico
plt.figure(figsize=(8, 6))  
plt.scatter(datos[:, 0], datos[:, 1], alpha=0.6, edgecolors='darkgreen', linewidths=0.7)
plt.title('Scatter plot', fontsize=14, fontweight='bold', color='Darkred')
plt.xlabel('X1', fontsize=12, color='Darkred')
plt.ylabel('X2', fontsize=12, color='Darkred')
plt.grid(True, linestyle='--', alpha=0.7)
plt.axis('equal')  
plt.axhline(0, color='gray',linewidth=0.5, ls='--')
plt.axvline(0, color='gray',linewidth=0.5, ls='--')
plt.savefig('diagrama_dispersión.png', dpi=400) 
plt.show()
```

#### 3. Visualización mediante histogramas

```python
# Definimos los rangos para las dimensiones
x_min = np.min(datos)
x_max = np.max(datos)
print(f"\nValor mínimo: {x_min}")
print(f"Valor máximo: {x_max}")
```

```python
# Creamos dos Histogramas usando los minimos y máximos
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(6, 10))

# Histograma para la primera dimensión
ax1.hist(datos[:, 0], bins=30, color='orange', edgecolor='black', alpha=0.7)
ax1.set_title('Histogram of X1', fontsize=14, fontweight='bold', color='Darkred')
ax1.set_xlabel('X1', fontsize=12, color='Darkred') 
ax1.set_ylabel('Frecuency', fontsize=12, color='Darkred')
ax1.grid(True, linestyle='--', alpha=0.7)
ax1.set_xlim(x_min, x_max)

# Histograma para la segunda dimensión
ax2.hist(datos[:, 1], bins=30, color='purple', edgecolor='black', alpha=0.7)
ax2.set_title('Histogram of X2', fontsize=14, fontweight='bold', color='Darkred')
ax2.set_xlabel('X2', fontsize=12, color='Darkred') 
ax2.set_ylabel('Frecuency', fontsize=12, color='Darkred')
ax2.grid(True, linestyle='--', alpha=0.7)
ax2.set_xlim(x_min, x_max)

plt.tight_layout()
plt.savefig('histogramas.png', dpi=400)
plt.show()
```

#### 4. Agregar datos

```python
# Generamos los nuevos datos aleatorios
media2 = [0, -3]
datos2 = np.random.multivariate_normal(media2, matriz_covarianza, size=500)

# Verificamos las dimensiones de los nuevos datos
print("\nDimensiones de los nuevos datos:", datos2.shape)
```

```python
# Configuramos el grafico de dispersión combinado
plt.figure(figsize=(10, 8))
plt.scatter(datos[:, 0], datos[:, 1], alpha=0.6, edgecolors='darkgreen', linewidths=0.7, label='0')
plt.scatter(datos2[:, 0], datos2[:, 1], alpha=0.6, edgecolors='darkred', linewidths=0.7, label='1')
plt.title('Combined Scatter Plot', fontsize=14, fontweight='bold', color='Darkred')
plt.xlabel('X1', fontsize=11, color='Darkred')
plt.ylabel('X2', fontsize=11, color='Darkred')
plt.grid(True, linestyle='--', alpha=0.7)
plt.axis('equal')
plt.axhline(0, color='gray',linewidth=0.5, ls='--')
plt.axvline(0, color='gray',linewidth=0.5, ls='--')
plt.legend() 
plt.savefig('diagrama_dispersión_combinado.png', dpi=400)
plt.show()
```

#### 5. Combinación de datos

```python
# Combianmos los array creados en uno solo
# opción 1
datos_combinados = np.vstack((datos, datos2))

# opción 2
datos_combinados2 = np.concatenate((datos, datos2), axis=0)

# Verificamps las dimensiones de los datos combinados
print("\nDimensiones de datos combinados con vstack:", datos_combinados.shape)
print("Dimensiones de datos combinados con concatenate:", datos_combinados2.shape)
```

#### 6. Etiquetado

```python
# Creamos las etiquetas para los datos combinados
etiqueta_datos1 = np.zeros(datos.shape[0], dtype=int)
etiqueta_datos2 = np.ones(datos2.shape[0], dtype=int)

# Combinamos las etiquetas
etiquetas_combinadas = np.concatenate((etiqueta_datos1, etiqueta_datos2))

# Verificamos las dimensiones de las etiquetas combinadas
print("\nDimensiones de las etiquetas combinadas:", etiquetas_combinadas.shape)
```

```python
# Creamos las etiquetas
etiqueta1 = np.zeros((500, 1))
etiqueta2 = np.ones((500, 1))

# Comibnamos las etiquetas
etiquetas = np.vstack((etiqueta1, etiqueta2))

# Verificamos las dimensiones de las etiquetas combinadas
print("\nDimensiones de las etiquetas combinadas (vstack):", etiquetas.shape)
```

```python
# Combinamos los datos combianados y las etiquetas en una sola matriz
datos_etiquetados = np.hstack((datos_combinados, etiquetas))

# Verificamos las dimensiones de la matriz final
print("\nDimensiones de la matriz final de datos etiquetados:", datos_etiquetados.shape)
print(datos_etiquetados)
```
