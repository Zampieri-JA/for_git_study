Cuenta la historia que cuando Toyotomi Hideyoshi le ofreció una recompensa, Shinzaemon le pidió lo siguiente: un grano de arroz el primer día, dos granos el siguiente, cuatro granos al día siguiente, y así sucesivamente, duplicando la cantidad cada día durante 100 días.

#### Cálculo del número de Arróz en el día 100

```python
# Importanos librería
import matplotlib.pyplot as plt
```

```python
# Datos iniciales
dias = 100
```

```python
# Creamos una lista  vacía
granos_por_dia = []  # Lista para guardar cuántos granos recibe cada día
total_acumulado = 0  # Almacena todos los granos recibidos

# Iniciamos el ciclo for para los granos por día
for dia in range(1, dias + 1):
    granos_hoy = 2 ** (dia - 1)  # cada día se duplica la cantidad de granos
    
    # Agremamos los valores a la lista y sumamos el total
    granos_por_dia.append(granos_hoy)
    total_acumulado += granos_hoy

# imprimimos los resultados
print(f"Granos recibidos el día 100: {granos_por_dia[99]:,}")
print(f"Total de granos acumulados en los 100 días: {total_acumulado:,}")
```

##### Graficando la cantidad de arroz
He tomado 5 días para la representación

```python
# Días elgidos para mostrar en la gráfica
dias_mostrar = [1, 25, 50, 75, 100]
granos_dias_especificos = []

# Calcular granos para cada día específico
for dia in dias_mostrar:
    granos = 2 ** (dia - 1)
    granos_dias_especificos.append(granos)
```

```python
# Configuramos una gráfica
plt.figure(figsize=(15, 8))
plt.plot(dias_mostrar, granos_dias_especificos, linewidth=3, markersize=10, color='darkgreen', 
         label='Cantidad de granos por día', marker='o')
plt.yscale('log')

# Etiquetas y título
plt.xlabel('Días', fontsize=14, fontweight='bold', color='darkred')
plt.ylabel('Cantidad de Granos de arroz', fontsize=14, fontweight='bold', color='darkred')
plt.title('Granos de Arroz Recibidos\n(Días 1, 25, 50, 75 y 100)', fontsize=15, fontweight='bold', color='darkred')

# Agregar valores
for i, (dia, granos) in enumerate(zip(dias_mostrar, granos_dias_especificos)):
    plt.annotate(f'{granos:,}', 
                xy=(dia, granos), 
                xytext=(-35, 25), 
                textcoords='offset points',
                fontsize=9,
                bbox=dict(boxstyle='round,pad=0.3', facecolor='orange', alpha=0.5),
                arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0'))

# Configurar la cuadrícula
plt.grid(True, alpha=0.3, linestyle='--', color='gray')
plt.legend()

# Agregar texto del total acumulado en la esquina inferior derecha
plt.text(0.38, 0.8, f'Total acumulado en 100 días:\n{total_acumulado:,} granos de arroz', 
         transform=plt.gca().transAxes,
         fontsize=12,
         verticalalignment='bottom',
         horizontalalignment='right',
         bbox=dict(boxstyle='round,pad=0.5', facecolor='orange', alpha=0.8),
         fontweight='bold')

# Ajustar el layout y mostrar
#plt.tight_layout()
plt.savefig('Granos_de_Arroz_Shinzaemon.png', dpi=400)
plt.show()

# Mostrar los valores en consola también
print("Granos de Arroz en los días especificados:")
for dia, granos in zip(dias_mostrar, granos_dias_especificos):
    print(f"Día {dia}: {granos:,} granos")

print(f"\nTotal acumulado en 100 Días: {total_acumulado:,} granos")
```
