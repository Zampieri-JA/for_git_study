El descenso de gradiente es una técnica que se utiliza a menudo para encontrar el mínimo de una función. Consideramos la siguiente ecuación donde $x$ varía de -50 a 50:



```python
# Importamos la librería numpy
import numpy as np
```

```python
# Definimos la ecuación
def equation(x):
    return (1/2) * x + 1
```

```python
# Creamos un ndarray que represente los valores de x de -50 a 50 en incrementos de 0,1
x_valores = np.arange(-50, 50.1, 0.1)

# calcule el valor de y correspondiente a cada x y guárdelo en el ndarray
y_valores = equation(x_valores)

# Mostramos los valores de x e y
print("Valores de x:\n", x_valores)
print("\nValores de y:\n", y_valores)
```

#### Concatenación de matrices

```python
# Combinamos array en uno solo
data1 = np.column_stack((x_valores, y_valores))
print("\nDatos combinados (x, y):\n", data1)
```

```python
# Mostramos la forma del array combinado
print("\nForma del array combinado:", data1.shape)
```

#### Encontrar el gradiente

```python
# Calculamso el cambio en Y y el cambio en X
cambio_y = np.diff(y_valores)
cambio_x = np.diff(x_valores)
```

```python
# Calculamos la pendiente (gradiente)
Gradiente = cambio_y / cambio_x

# Mostramos la pendiente
print("\nPendiente (gradiente):\n", pendiente.shape)
print(Gradiente)
```

#### Dibujar un gráfico

```python
# Para graficar la gradiente, creamos un nuevo array de x que es un elemento menos que el original
gradiente_plot = np.full_like(x_valores, Gradiente[0])

# Creamos el gráfico
plt.figure(figsize=(10, 6))
plt.plot(x_valores, y_valores, label='Ecuación: y = (1/2)x + 1', color='darkblue', linewidth=3)
plt.plot(x_valores[:-1], gradiente_plot[:-1], label='Gradiente', color='red', linestyle='--')
plt.title('Ecuación y su Gradiente', fontsize=14, color='darkgreen') 
plt.xlabel('X', fontsize=12, color='darkgreen')
plt.ylabel('Y', fontsize=12, color='darkgreen')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()

# Los ejes en el centro para una mejor visualización
ax = plt.gca()
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.savefig('ecuacion_gradiente.png', dpi=400) 
plt.show()
```

#### Funciones de Python

Calcule la pendiente de las siguientes tres ecuaciones en incrementos de 0,1:



```python
# Creamos una función para analizar y graficar cualquier función.
def analyze_function(func, x_range_start, x_range_end, step=0.1, title=""):
    """
    Calcula los valores de y y el gradiente de una función,
    y luego grafica la función y su gradiente.

    Args:
        func (function): La función matemática a analizar.
        x_range_start (int): El valor inicial del rango de x.
        x_range_end (int): El valor final del rango de x.
        step (float): El incremento para los valores de x.
        title (str): El título para el gráfico.
    """
    # Creamos nuevamente los valores de x y, calculamos y
    x_values = np.arange(x_range_start, x_range_end + step, step)
    y_values = func(x_values)

    # Calculamos la gradiente
    delta_y = np.diff(y_values)
    delta_x = np.diff(x_values)
    gradient = delta_y / delta_x
    
    # Creamos la gráfica
    plt.figure(figsize=(10, 6))

    # graficamos la función original
    plt.plot(x_values, y_values, label=f'{title}', color='blue', linewidth=2)

    # Graficamos el gradiente
    plt.plot(x_values[:-1], gradient, label='Gradiente', color='red', linestyle='--', linewidth=2)

    # Completamos el gráfico
    plt.title(f'Gráfico de la Ecuación y su Gradiente: {title}', fontsize=14, color='darkgreen')
    plt.xlabel('X', fontsize=12, color='darkgreen')
    plt.ylabel('Y', fontsize=12, color='darkgreen')
    plt.grid(True)
    plt.legend()
    safe_title = re.sub(r'[^\w\s]', '', title).strip()
    file_name = f'grafico_{safe_title.replace(" ", "_")}.png'
    plt.savefig(file_name, dpi=400)
    plt.show()

# Analizamos las Ecuaciones dadas
def func1(x):
    return x**2 #Ecuación 1: y = x^2

def func2(x):
    return 2 * x**2 + 2**x  #Ecuación 2: y = 2x^2 + 2^x

def func3(x):
    return np.sin(np.sqrt(x))   #Ecuación 3: y = sin(x^(1/2))

# Agramos las gradientes

analyze_function(func1, -50, 50, title='$y = x^2$')
analyze_function(func2, -50, 50, title='$y = 2x^2 + 2^x$')
analyze_function(func3, 0, 50, title='$y = \sin(\sqrt{x})$')
```

#### Encuentra el valor mínimo

Halle el valor mínimo de y para cada función del Problema anterior

```python
# Definimos una nueva función para encontrar el valor mínimo y su gradiente
def func1(x):
    return x**2

def func2(x):
    return 2 * x**2 + 2**x

def func3(x):
    return np.sin(np.sqrt(x))

def find_minima(func, x_range_start, x_range_end, step=0.1, title=""):
    """
    Calcula el valor mínimo de una función y su gradiente.

    Args:
        func (function): La función matemática a analizar.
        x_range_start (int): El valor inicial del rango de x.
        x_range_end (int): El valor final del rango de x.
        step (float): El incremento para los valores de x.
        title (str): El título de la función.
    """
    # creamos los valores de X y Y
    x_values1 = np.arange(x_range_start, x_range_end + step, step)
    y_values1 = func(x_values1)

    # Obtenemos la gradiente
    delta_y1 = np.diff(y_values1)
    delta_x1 = np.diff(x_values1)
    gradient = delta_y1 / delta_x1
    
    # Obtenemos el valor mínimo
    min_y_value = np.min(y_values1)
    min_y_index = np.argmin(y_values1)
    
    print(f"Ecuación: {title} ---")
    print(f"Valor mínimo de Y: {min_y_value}")
    print(f"Índice del valor mínimo: {min_y_index}")
    
    # Mostramos la gradiente antes y después del valor mínimo
    if min_y_index > 0:
        print(f"Gradiente antes del valor mínimo de Y: {gradient[min_y_index - 1]}")
    if min_y_index < len(gradient):
        print(f"Gradiente después del valor mínimo de Y: {gradient[min_y_index]}")
    
    print("\n")

# Ejecutamos las funciones para cada una de las tres ecuaciones
find_minima(func1, -50, 50, title='$y = x^2$')
find_minima(func2, -50, 50, title='$y = 2x^2 + 2^x$')
find_minima(func3, 0, 50, title='$y = \sin(\sqrt{x})$')
```
