Si en un tablero de ajedrez se colocara trigo en cada casilla de manera que se colocara un grano en la primera casilla, dos en la segunda, cuatro en la tercera y así sucesivamente (duplicando el número de granos en cada casilla subsiguiente)

#### ¿Cuántos granos de trigo habría en el tablero de ajedrez al final?

Este problema lo abordo primero mostrando un ejemplo usando numpy con un Array 4x4, tomo .zeros() en lugar de .array() para crear el array

```python
# Importamos Numpy
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
```

```python
# Creamos un array de 16 elementos para representar las casillas de un tablero 4x4
tablero1 = np.zeros(16, dtype=np.int64)
```

```python
# Llenamos el array con la cantidad de granos en cada casilla
for i in range(16):
    tablero1[i] = 2**i
```

```python
# Verificamos el tablero 4x4
print(tablero1.reshape(4, 4))
```

```python

# Calculamos el total de granos usando numpy.sum()
total_granos_4x4 = np.sum(tablero1)

# Mostramos el resultado
print("Número total de granos en el tablero de 4x4:", total_granos_4x4)
```

Respondiendo a la pregunta del problema, usamos el ejemplo anterior

```python
# Creammos el tablero de ajedrez 8x8
tablero_ajedrez = 64    # casillas
```

```python
# Creamos el array de 8x8
tablero = np.zeros(tablero_ajedrez, dtype=object)
```

```python
# Creamos el ciclo for para lleanar el tablero
for i in range(tablero_ajedrez):
    tablero[i] = 2**i
```

```python
# Verificamos el tablero
print(tablero.reshape(32, 2))
```

```python
# Calculamos la sumatorio de todos los granos de trigo
total_granos_trigo = np.sum(tablero)

# Mostramos el resultado
print(f"Total de granos de trigo en el tablero de ajedrez: {total_granos_trigo}")
```

#### Expansión a n×m cuadrículas

```python
# Creamos una función para calcular los granos en un tablero n x m
def calcular_granos_tablero(n, m):
    """
    Calcula el número de granos de trigo en el tablero de ajedrez de n x m,
    donde la cantidad de granos se duplica en cada casilla.

    Args:
        n (int): El número de filas del tablero.
        m (int): El número de columnas del tablero.

    Returns:
        numpy.ndarray: Un array 1D con la cantidad de granos en cada casilla.
    """
    total_casillas = n * m
    
    # Creamos un array vacío para almacenar los granos.
    granos = np.zeros(total_casillas, dtype=object)
    
    # Llenamos el array con la cantidad de granos en cada casilla
    for i in range(total_casillas):
        granos[i] = 2**i
        
    return granos
```

```python
# Aquí se puede modificar el tamaño del tablero
filas = 8
columnas = 8

# Llamamos a la función
tablero_funcion = calcular_granos_tablero(filas, columnas)

# Mostramos el tablero
print(tablero_funcion.reshape(filas, columnas))

# Calculamos el total de gramos en el tablero
granos_totales = np.sum(tablero_funcion)
print("\n")
print(f"Número total de granos en el tablero: {granos_totales}")
```

#### Visualizando los promedios por columnas

```python
# Calculamos el promedio de granos por columna
mean_columns = np.mean(tablero_funcion.reshape(filas, columnas), axis=0)

# Imprmimos el promedio por columnas mediante un ciclo for
for i, promedio in enumerate(mean_columns):
    print(f"Promedio de granos en la columna {i+1}: {promedio:,.0f}")
```

```python
# Generamos una lista con los nombres de las columnas
nombres_columnas = [f'Columna {i+1}' for i in range(columnas)]

plt.figure(figsize=(12, 7))
plt.bar(nombres_columnas, mean_columns, color='skyblue')

# Usamos una escala logarítmica para el eje Y para mostrar el crecimiento exponencial
plt.yscale('log')
plt.title('Promedio de Granos de Trigo por Columna', fontsize=12, fontweight='bold', color='darkblue')
plt.xlabel('Número de Columna', fontsize=10, fontweight='bold', color='darkblue')
plt.ylabel('Promedio de Granos de Trigo', fontsize=10, fontweight='bold', color='darkblue')
plt.xticks(rotation=45) # Rotamos las etiquetas para mayor legibilidad
plt.grid(axis='y', linestyle='--', alpha=0.5) # Añadimos una cuadrícula
plt.tight_layout() 
plt.savefig('promedio_por_columna.png', dpi=400) 
plt.show()
```

Mediante un Mapa de calor

```python
# convertimos de objecto a float para poder graficar
tablero_funcion1 = tablero_funcion.astype(np.float64)
tablero_funcion2 = tablero_funcion1.reshape(filas, columnas)
```

```python
# --- 2. Visualizar con un mapa de calor ---
plt.figure(figsize=(9, 8))

# Usamos imshow para crear el mapa de calor.
# El argumento norm=LogNorm() aplica una escala logarítmica.
mapa_de_calor = plt.imshow(tablero_funcion2, cmap='plasma', norm=LogNorm(vmin=1, vmax=tablero_funcion2.max()))

# Agregamos una barra de color para la leyenda
plt.colorbar(mapa_de_calor, label='Número de Granos en el tablero')

# Configuramos etiquetas para los ejes
plt.title('Heatmap', fontsize=14, fontweight='bold', color='darkblue')
plt.xlabel('Columns', fontsize=10, fontweight='bold', color='darkblue')
plt.ylabel('Rows', fontsize=10, fontweight='bold', color='darkblue')
plt.xticks(np.arange(columnas), labels=np.arange(1, columnas + 1))
plt.yticks(np.arange(filas), labels=np.arange(1, filas + 1))
plt.savefig('mapa_de_calor.png', dpi=400)
plt.show()
```

¿Cuántas veces la segunda mitad es más grande que la primera?

```python
# Realizamos la suma de la primera mitad
p_mitad = tablero_funcion2[0:4, :]
sum_p_mitad = np.sum(p_mitad)

# Realizamos la suma de la segunda mitad
s_mitad = tablero_funcion2[4:8, :]
sum_s_mitad = np.sum(s_mitad)

# Calculamos el resultado
resultado = sum_s_mitad / sum_p_mitad
print("\n")
print(f"La Segunda mitad es: {resultado:.0f} veces es más grande que la primera mitad.")
```
